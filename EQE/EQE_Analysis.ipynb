# EQE Analysis and Jsc Calculation

This notebook analyzes the External Quantum Efficiency (EQE) data and calculates the short-circuit current density (Jsc) using the AM1.5G solar spectrum.

## Formula
$$J_{sc} = \frac{q}{hc} \int_{\lambda_1}^{\lambda_2} EQE(\lambda) \times \lambda \times \Phi_{AM1.5G}(\lambda) \, d\lambda$$

where:
- **EQE(λ)**: External Quantum Efficiency
- **Φ_AM1.5G(λ)**: AM1.5G solar spectrum irradiance  
- **λ**: Wavelength
- **Integration range**: 300 - 1100 nm

This program takes the data from the EQE spectra, multiply it for the respective AM1.5G spectra and take the numerical integral by applying the simpson's rule

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate, integrate
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
def interpolate_values(x, y, new_x):
    """
    Interpolate y values at new x positions using linear interpolation.
    """
    f = interpolate.interp1d(x, y, kind='linear', fill_value='extrapolate')
    return f(new_x)

def calculate_jsc(eqe_data, am15g_spectrum):
    """
    Calculate short-circuit current density (Jsc) from EQE data and AM1.5G spectrum.
    
    Parameters:
    -----------
    eqe_data : DataFrame
        Must have columns 'WL' (wavelength in nm) and 'QE' (quantum efficiency)
    am15g_spectrum : DataFrame
        Must have columns 'WL' (wavelength in nm) and 'AM1.5G' (irradiance in W/m²/nm)
    
    Returns:
    --------
    Jsc : float
        Short-circuit current density in mA/cm²
    """
    df_eqe = eqe_data.copy()
    df_am15 = am15g_spectrum.copy()
    
    # Convert QE to fraction if it's in percentage
    if df_eqe['QE'].mean() > 1:
        df_eqe['QE'] = df_eqe['QE'] / 100.0
    
    # Set index and crop AM1.5G spectrum to EQE wavelength range
    df_eqe.set_index('WL', inplace=True)
    df_am15.set_index('WL', inplace=True)
    df_am15 = df_am15[(df_am15.index >= df_eqe.index.min()) & (df_am15.index <= df_eqe.index.max())]
    
    # Interpolate EQE to match AM1.5G wavelength grid
    f_eqe = interpolate.interp1d(df_eqe.index, df_eqe['QE'], kind='linear', fill_value='extrapolate')
    df_am15['EQE_interp'] = f_eqe(df_am15.index)
    
    # Calculate integrand: EQE(λ) × AM1.5G(λ) × λ / (hc)
    # hc = 1239.84193 eV·nm
    df_am15['integrand'] = df_am15['AM1.5G'] * df_am15.index / 1239.84193 * df_am15['EQE_interp']
    
    # Integrate using Simpson's rule and convert to mA/cm²
    jsc = integrate.simpson(df_am15['integrand'], x=df_am15.index) / 10.0
    
    return jsc

print("Helper functions defined successfully!")
```

## 5.2 Generate Corrected Results Excel File

### Parameter Dependencies on Jsc Correction Factor

The following parameters are corrected based on their relationship to the Jsc value determined from the EQE measurement:

#### **1. Short-Circuit Current Density (Jsc)**
$$J_{sc,corrected} = J_{sc,original} \times CF$$

- **Direction**: Multiplied by correction factor
- **Reason**: Directly proportional to the EQE-derived Jsc. If the correction factor > 1, the original Jsc was underestimated.

#### **2. Fill Factor (FF)**
$$FF_{corrected} = \frac{FF_{original}}{CF}$$

- **Direction**: Divided by correction factor
- **Reason**: Inversely related to Jsc correction. When Jsc is increased by the correction factor, the FF must be adjusted inversely to maintain consistency with the original J-V curve characteristics (the curve shape determines FF independently of absolute current values).

#### **3. Power Conversion Efficiency (PCE/η)**
$$\eta_{corrected} = \eta_{original} \times CF$$

- **Direction**: Multiplied by correction factor
- **Reason**: Direct product of Jsc (among other factors). Since PCE ∝ Jsc, the efficiency scales proportionally with the Jsc correction factor.

#### **4. Open-Circuit Voltage (Voc)**
$$V_{oc,corrected} = V_{oc,original}$$

No correction applied.

- **Direction**: Unchanged
- **Reason**: Voc depends primarily on the bandgap and recombination properties of the material, not on the absolute photocurrent value. It remains constant regardless of Jsc scaling.

---

**Note**: The correction factor is calculated as:

$$CF = \frac{J_{sc,EQE}}{J_{sc,solar simulator}}$$

```python
# Create corrected results table
corrected_df = results_df.copy()

# Find column indices for Voc, Jsc, FF (Fill Factor), and PCE (case-insensitive)
col_names = corrected_df.columns.str.lower()

voc_col = None
jsc_col_corr = None
ff_col = None
pce_col = None

for col in corrected_df.columns:
    col_lower = col.lower()
    if 'voc' in col_lower and voc_col is None:
        voc_col = col
    elif ('jsc' in col_lower or ('j' in col_lower and 'sc' in col_lower)) and jsc_col_corr is None:
        jsc_col_corr = col
    elif ('fill factor' in col_lower or 'ff' in col_lower) and ff_col is None:
        ff_col = col
    elif ('pce' in col_lower or 'efficiency' in col_lower or 'eta' in col_lower) and pce_col is None:
        pce_col = col

print("Detected columns for correction:")
print(f"  Voc column: {voc_col}")
print(f"  Jsc column: {jsc_col_corr}")
print(f"  FF (Fill Factor) column: {ff_col}")
print(f"  PCE (Efficiency) column: {pce_col}")

# Apply corrections
if jsc_col_corr is not None:
    corrected_df[jsc_col_corr] = corrected_df[jsc_col_corr] * correction_factor
    print(f"\n✓ Corrected Jsc: multiplied by {correction_factor:.6f}")

if ff_col is not None:
    corrected_df[ff_col] = corrected_df[ff_col] / correction_factor
    print(f"✓ Corrected Fill Factor: divided by {correction_factor:.6f}")

if pce_col is not None:
    corrected_df[pce_col] = corrected_df[pce_col] * correction_factor
    print(f"✓ Corrected PCE (Efficiency): multiplied by {correction_factor:.6f}")

# Select only the relevant columns (Sample, Voc, Jsc, Fill Factor, PCE)
columns_to_keep = [corrected_df.columns[0]]  # First column (sample name)
if voc_col is not None:
    columns_to_keep.append(voc_col)
if jsc_col_corr is not None:
    columns_to_keep.append(jsc_col_corr)
if ff_col is not None:
    columns_to_keep.append(ff_col)
if pce_col is not None:
    columns_to_keep.append(pce_col)

corrected_results = corrected_df[columns_to_keep].copy()

# Add Correction Factor column 2 cells after Efficiency
corrected_results[''] = ''  # Blank column
corrected_results['Correction Factor'] = correction_factor

# Save to Excel
output_file = 'C:/Users/jp_ol/OneDrive/Ambiente de Trabalho/TESE/Data_Analysis/data/CdS3_1/CdS3_1_Results_Table_CORRECTED.xlsx'

try:
    with pd.ExcelWriter(output_file, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
        corrected_results.to_excel(writer, sheet_name='Corrected Results', index=False)
    print(f"\n" + "="*70)
    print(f"CORRECTED RESULTS TABLE GENERATED")
    print(f"="*70)
    print(f"\nFile saved: {output_file}")
except PermissionError:
    print(f"\nWARNING: Could not save Excel file - it may be open in another application.")
    print(f"Path: {output_file}")
    print(f"\nResults (not saved to Excel):")
    print(corrected_results)
```
